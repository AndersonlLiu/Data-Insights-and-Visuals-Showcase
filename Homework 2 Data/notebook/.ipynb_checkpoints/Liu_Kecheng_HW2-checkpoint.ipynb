### 1. i) KNN Regression

```python
# Split data into 80% train and 20% test
train_data, test_data = train_test_split(df, test_size=0.2)
features = ['AT', 'V', 'AP', 'RH']
train_X = train_data[features]
train_y = train_data['PE']
test_X = test_data[features]
test_y = test_data['PE']
```

```python
#knn regression with raw feature
BestN = 0
LowestError = 1
TestErrorList = []
TrainErrorList = []
for n in range(1, 101):
    neigh = KNeighborsRegressor(n_neighbors = n)
    neigh.fit(train_X, train_y)
    test_error = 1 - neigh.score(test_X, test_y)
    if test_error < LowestError:
        LowestError = test_error
        BestN = n
    TestErrorList.append(test_error)
    TrainErrorList.append(1 - neigh.score(train_X, train_y))
print('Lowest Error is', LowestError)
print('Best N is', BestN)
```

```python
# print out the test MSE for knn at the best N
neigh_mse = KNeighborsRegressor(n_neighbors = 6)
neigh_mse.fit(train_X, train_y)
test_mse_knn = mean_squared_error(test_y, neigh_mse.predict(test_X))
print('The test MSE is', test_mse_knn)
```

```python
#Reverse the errorlist to fit the 1/k graph
TestErrorList.reverse()
TrainErrorList.reverse()
```

```python
#ploting the train and test error
plt.figure(figsize=(12, 6))
plt.plot(np.arange(0.01, 1.01, 0.01), TestErrorList, color='red', linestyle='dashed', marker='o',
         markerfacecolor='black', markersize=9)
plt.plot(np.arange(0.01, 1.01, 0.01), TrainErrorList, color='blue', linestyle='dashed', marker='o',
         markerfacecolor='black', markersize=9)
plt.title('Test Error & Train Error vs. 1/k')
plt.xlabel('1/k')
plt.ylabel('Error')
```

```python
#Normalized data
x = df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_new = pd.DataFrame(x_scaled)
df_new
```

```python
# Split data into 80% train and 20% test for normalized data
train_data_Normalized, test_data_Normalized = train_test_split(df_new, test_size=0.2)
features = [0, 1, 2, 3]
train_X_Normalized = train_data_Normalized[features]
train_y_Normalized = train_data_Normalized[4]
test_X_Normalized = test_data_Normalized[features]
test_y_Normalized = test_data_Normalized[4]
```

```python
#knn regression with normalized feature
BestN_Normalized = 0
LowestError_Normalized = 1
TestErrorList_Normalized = []
TrainErrorList_Normalized = []
for n in range(1, 101):
    neigh = KNeighborsRegressor(n_neighbors = n)
    neigh.fit(train_X_Normalized, train_y_Normalized)
    test_error = 1 - neigh.score(test_X_Normalized, test_y_Normalized)
    if test_error < LowestError_Normalized:
        LowestError_Normalized = test_error
        BestN_Normalized = n
    TestErrorList_Normalized.append(test_error)
    TrainErrorList_Normalized.append(1 - neigh.score(train_X_Normalized, train_y_Normalized))
print('Lowest Error for normalized features is', LowestError_Normalized)
print('Best N for normalized feature is', BestN_Normalized)
```

```python
# print out the test MSE AFTER NORMALIZED for knn at the best N
neigh_mse_normalized = KNeighborsRegressor(n_neighbors = 4)
neigh_mse_normalized.fit(train_X_Normalized, train_y_Normalized)
test_mse_knn_normalized = mean_squared_error(test_y_Normalized, neigh_mse_normalized.predict(test_X_Normalized))
print('The test MSE after normalized is', test_mse_knn_normalized)
```

```python
#Reverse the errorlist to fit the 1/k graph
TestErrorList_Normalized.reverse()
TrainErrorList_Normalized.reverse()
```

```python
#ploting the train and test error for normalized data
plt.figure(figsize=(12, 6))
plt.plot(np.arange(0.01, 1.01, 0.01), TestErrorList_Normalized, color='red', linestyle='dashed', marker='o',
         markerfacecolor='black', markersize=9)
plt.plot(np.arange(0.01, 1.01, 0.01), TrainErrorList_Normalized, color='blue', linestyle='dashed', marker='o',
         markerfacecolor='black', markersize=9)
plt.title('Test Error & Train Error vs. 1/k for normalized data')
plt.xlabel('1/k')
plt.ylabel('Error')
```

### 2.4.1 . For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

#### a) When the sample size is large and the number of predictors is small, we may underfit the data if we decided to set a restricted assumption on it. On the contrary, using a flexible method would reduce the bias we get, and has a more accurate result.

#### b) When the number of predictors is large and the number of observations is small, using inflexible method would be better. Due to the small sample size, using a too flexible method may cause overfitting. 

#### c) When the relationship is highly-nonlinear, it would be better using the inflexible method. Becasue non-linear machine learning model has low bias but high variance, we need to use a flexible method to find the correct fit

#### d) When the variance of the error term is extremely high, it's better to use inflexible method. Because a flexible method has a larger variance which would including a lot of noise terms.
