```python
# Company News
from urllib.request import urlopen
from urllib.request import Request
from bs4 import BeautifulSoup
finviz_url = 'https://finviz.com/quote.ashx?t=AAPL'
news_tables = {}
req = Request(url=finviz_url,headers={'user-agent': 'my-app/0.0.1'}) 
resp = urlopen(req)    
html = BeautifulSoup(resp, features="lxml")
news_table = html.find(id='news-table')
news_tables['AAPL'] = news_table
```

```python
df = news_tables['AAPL']
df_tr = df.findAll('tr')
print ('\n')
print ('Recent News Headlines for {}: '.format('AAPL'))
for i, table_row in enumerate(df_tr):
    a_text = table_row.a.text
    td_text = table_row.td.text
    td_text = td_text.strip()
    print(a_text,'(',td_text,')')
```

```python
parsed_news = []
for file_name, news_table in news_tables.items():
    for x in news_table.findAll('tr'):
        text = x.a.get_text() 
        date_scrape = x.td.text.split()

        if len(date_scrape) == 1:
            time = date_scrape[0]
            
        else:
            date = date_scrape[0]
            time = date_scrape[1]

        ticker = file_name.split('_')[0]
        
        parsed_news.append([ticker, date, time, text])
```

```python
from datetime import datetime
news = pd.DataFrame(parsed_news, columns = ['Stock', 'Date', 'Time', 'News Headlines'])
news['Date'] = pd.to_datetime(news['Date'], format='%b-%d-%y')
```

### Stationary Test for pct_change (Used to test whether the time series analysis is meaningful)

```python
# Time Series plot of the pct_change
plt.figure(figsize=(16,5))
plt.plot(data_SERIES)
```

```python
from statsmodels.tsa.stattools import adfuller
data_new = stock_price[['date', 'pct_change']].set_index(['date'])
def test_stationarity(timeseries):
    
    # Calculate rolling mean and sd
    rolmean = timeseries.rolling(10).mean()
    rolstd = timeseries.rolling(10).std()

    # Plot the rolling stat
    plt.figure(figsize=(24, 8))   
    orig = plt.plot(timeseries, color='blue',label='Original') 
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std') 
    
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # ADF Test (stationarity test)
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used',
                                             'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
test_stationarity(data_new)
```

```python
# White Noise Test
from statsmodels.stats.diagnostic import acorr_ljungbox
print('White Noise：',acorr_ljungbox(data_SERIES,lags=1))
```

### Stationary Test for close price

```python
data_closePrice = stock_price[['date', 'close_price']].set_index(['date'])
test_stationarity(data_closePrice)
```

### The P-value is too large to reject the null hypothesis (>> 0.05), So we can't apply the ARIMA model yet

```python
from math import sin
from math import radians
from matplotlib import pyplot
 
close = data_closePrice['close_price'].tolist()
# create a differenced series
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return diff
 
# invert differenced forecast
def inverse_difference(last_ob, value):
    return value + last_ob

# difference the dataset
diff = difference(close, 20)
pyplot.plot(diff)
pyplot.show()

# invert the difference
inverted = [inverse_difference(close[i], diff[i]) for i in range(len(diff))]
pyplot.plot(inverted)
pyplot.show()
```

```python
# ADF Test (stationarity test)
print('Results of Dickey-Fuller Test:')
dftest = adfuller(diff, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used',
                                         'Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
print('White Noise：',acorr_ljungbox(diff,lags=1))
```

### The result shows that the difference is a stable time series data and is not white noise. So it's worth further modeling :)

```python
# Preparing data for time series
date = stock_price['date'].tolist()[20:]
d = {'date': date,'price_diff': diff}
df = pd.DataFrame(d)
df = df.set_index(['date'])
```

```python
plot_acf(df)
plot_pacf(df)
plt.show()
```
