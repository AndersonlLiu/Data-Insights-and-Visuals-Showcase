```python
def one_zero_loss(y_true, y_pred):
    n = len(y_true)
    row_indicators = np.logical_not(np.all(y_true == y_pred, axis = 1)) # axis = 1 will check for equality along rows.
    not_equal_count = np.sum(row_indicators)
    return not_equal_count/n
# Resources: https://medium.datadriveninvestor.com/a-survey-of-evaluation-metrics-for-multilabel-classification-bb16e8cd41cd
```

```python
def emr(y_true, y_pred):
    n = len(y_true)
    row_indicators = np.all(y_true == y_pred, axis = 1) # axis = 1 will check for equality along rows.
    exact_match_count = np.sum(row_indicators)
    return exact_match_count/n
#Resources: https://medium.datadriveninvestor.com/a-survey-of-evaluation-metrics-for-multilabel-classification-bb16e8cd41cd
```

### K_Means

```python
# Resources: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html
# The below is for the test
X = data.iloc[:, 0:22]
y = data[['Family','Genus','Species']].to_numpy()
silhouette = {}
for k in range(2,51):
    kmeans = KMeans(random_state = 42, n_clusters = k).fit(X)
    #cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette.update({k: silhouette_avg})
best_k = max(silhouette, key=silhouette.get)
```

```python
best_k = max(silhouette, key=silhouette.get)
kmeans_new = KMeans(random_state = 42, n_clusters = best_k).fit(X)
labels = kmeans_new.labels_
data_predict = data.copy()
data_predict['predict_labels'] = labels.tolist()
```

```python
predict = data_predict['predict_labels'].to_list()

for i in range(0,best_k):
    label_data = data_predict.loc[data_predict['predict_labels'] == i]
    most_Family = label_data.Family.mode()[0]
    most_Genus = label_data.Genus.mode()[0]
    most_Species = label_data.Species.mode()[0]
    label = [most_Family, most_Genus, most_Species]
#   print(predict[0])
    predict = [label if item == i else item for item in predict]
```

```python
predict = np.array(predict, dtype = 'object')
hamming_loss = np.sum(np.not_equal(y, predict))/float(predict.size)
hamming_loss
############################################################################################
```

```python
hammingLose = []
X = data.iloc[:, 0:22]
y = data[['Family','Genus','Species']].to_numpy()
for time in range(1,51):
    silhouette = {}
    for k in range(2,51):
        kmeans = KMeans(n_clusters = k).fit(X)
        silhouette_avg = silhouette_score(X, kmeans.labels_)
        silhouette.update({k: silhouette_avg})
    best_k = max(silhouette, key=silhouette.get)
    kmeans_new = KMeans(n_clusters = best_k).fit(X)
    labels = kmeans_new.labels_
    data_predict = data.copy()
    data_predict['predict_labels'] = labels.tolist()
    predict = data_predict['predict_labels'].to_list()
    for i in range(0,best_k):
        label_data = data_predict.loc[data_predict['predict_labels'] == i]
        most_Family = label_data.Family.mode()[0]
        most_Genus = label_data.Genus.mode()[0]
        most_Species = label_data.Species.mode()[0]
        label = [most_Family, most_Genus, most_Species]
        predict = [label if item == i else item for item in predict]
    predict = np.array(predict, dtype = 'object')
    hamming_loss = np.sum(np.not_equal(y, predict))/float(predict.size)
    hammingLose.append(hamming_loss)
```

```python
import statistics
from statistics import mean
hammingDistance = [element * 3 for element in hammingLose]
hammingScore = [1 - element for element in hammingLose]
print(f'the std for hamming distance is {statistics.pstdev(hammingDistance)} and mean is {mean(hammingDistance)}')
```
